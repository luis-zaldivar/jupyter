```python
import random
from random import randint

def Conversion(decimal):
    binario=""
    while decimal>0:
        binario+=str(decimal%2)
        decimal=int(decimal/2)
    binario=binario[::-1]
    return binario

def ConvertirColumnaABinario(tabla):
    i=0
    while i<10:
        tabla[4][i]=Conversion(int(tabla[2][i]))
        i+=1

def f(x,y):
  return (x**2*(y))+(y**2*(x))

#def f2(x,y):
 #   return (x**2(y))+(y**2(x))

def VerificarRepeticiones(z,ArregloZ):
  #Se recorre todo el arreglo buscado el valor de z para ver si no esta repetido
  #Se regresa true si ya fue calculado y false si no
  for Actual in ArregloZ:
    if Actual==z: return True #si esto pasa significa que el valor ya existia en el arreglo
  return False #El valor no existe en el arreglo y retorna false

def GenerarAleatoriosYCalcularZ():
  
  #Generamos un arreglo de valores para y e x y z
  ArregloX=[]
  ArregloY=[]
  ArregloZ=[]
  ArregloGrado=[]
  ArregloBinario=[]
  #Para contar 10 iteraciones
  contador=1
  while contador<=10:
    #Generamos dos candidatos (uno para x y uno para y)
    x=randint(1,100)
    y=randint(1,100)
    z=f(x,y) #Calculamos el resultado al evaluar F(x,y)
    if VerificarRepeticiones(z,ArregloZ): 
      continue #Regrese al inicio del while
    
    ArregloX.append(x)
    ArregloY.append(y)
    ArregloZ.append(z)
    ArregloGrado.append(-1)
    ArregloBinario.append(-1)
    contador+=1 #Se aumenta el contador
  tabla=[]
    
  tabla.append(ArregloX)
  tabla.append(ArregloY)
  tabla.append(ArregloZ)
  tabla.append(ArregloGrado)
  tabla.append(ArregloBinario)
  return tabla #Regresamos la tabla

def ImprimirTabla(tabla):
    print("x\t\ty\t\tf(x,y)\t\tGrado\t\tBinario")
    for i in range(10):
        print(str(tabla[0][i])+"\t\t",end="")
        print(str(tabla[1][i])+"\t\t",end="")
        print(str(tabla[2][i])+"\t\t",end="")
        print(str(tabla[3][i])+"\t\t",end="")
        print(str(tabla[4][i]))
    print()

def Intercambiar(v1,v2):
    aux=v2
    v2=v1
    v1=aux
        
def SelectionSort(tabla):
    #Ordenar con el metodo de selection sort
    for i in range(10):
        valorMaximo=0
        posicionMaximo=0
        j=i
        while j<10:
            if tabla[2][j]>valorMaximo:
                valorMaximo=tabla[2][j]
                posicionMaximo=j
            j+=1
        tabla[0][i],tabla[0][posicionMaximo]=tabla[0][posicionMaximo],tabla[0][i]
        tabla[1][i],tabla[1][posicionMaximo]=tabla[1][posicionMaximo],tabla[1][i]
        tabla[2][i],tabla[2][posicionMaximo]=tabla[2][posicionMaximo],tabla[2][i]

def CalcularGrado(tabla):
    sumatoria=0
    for valor in tabla[2]:
        sumatoria+=valor
    i=0   
    while i<10:
        tabla[3][i]=tabla[2][i]/sumatoria
        i=i+1
    return sumatoria

def Mezclar(p,m):
    h1=""
    h2=""
    retorno=[]
    indice=0
    while indice<2 and indice<len(p):
        h1+=p[indice]
        indice+=1
    while indice<len(m):
        h1+=m[indice]
        indice+=1
        
    indice=0
    while indice<2 and indice<len(m):
        h2+=m[indice]
        indice+=1
    while indice<len(p):
        h2+=p[indice]
        indice+=1
    retorno.append(h1)
    retorno.append(h2)
    return retorno

def GenerarHijos(tabla):
    parejas=[]
    cont=0
    while cont<4:
        nuevaPareja=[]
        p=randint(0,9)
        m=randint(0,9)
        while m==p:
            m=randint(0,9)
        i=0
        while i<cont:
            if parejas[0][i]==p and parejas[1][i]==m:
                continue
            i+=1
        nuevaGeneracion=Mezclar(tabla[4][p],tabla[4][m])
        if nuevaGeneracion[0]==tabla[4][p] or nuevaGeneracion[0]==tabla[4][m] or nuevaGeneracion[1]==tabla[4][p] or nuevaGeneracion[1]==tabla[4][m]:
            continue
        print("padre:\t"+str(tabla[4][p]))
        print("madre:\t"+str(tabla[4][m]))
        print("Hijo 1:\t"+str(nuevaGeneracion[0]))
        print("Hijo 2:\t"+str(nuevaGeneracion[1]))
        print()
        nuevaPareja.append(p)
        nuevaPareja.append(m)
        parejas.append(parejas)
        cont+=1
    

tabla=GenerarAleatoriosYCalcularZ()
ImprimirTabla(tabla)
SelectionSort(tabla)
ImprimirTabla(tabla)
sumatoria=CalcularGrado(tabla)
ImprimirTabla(tabla)
print("sumatoria= "+str(sumatoria))
ConvertirColumnaABinario(tabla)
ImprimirTabla(tabla)
GenerarHijos(tabla)
```
